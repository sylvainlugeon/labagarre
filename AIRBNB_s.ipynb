```python
import pandas as pd
import numpy as np
from math import sin, cos, sqrt, atan2, radians
```

```python
#Constants
R = 6373.0 # earth's radius
```

```python
data = pd.read_csv('../data/AB_NYC_2019.csv')
data_sampled = data.sample(n=100)
```

```python
#data.head()
data_sampled.sort_values(by=['id'], inplace=True)
data_sampled.head()
#print(data_sampled)
```

```python
def measure_distance(lat0, lon0, lat1, lon1) :
    
    lat0 = radians(lat0)
    lon0 = radians(lon0)
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    
    dlat = lat1 - lat0
    dlon = lon1 - lon0

    a = sin(dlat / 2)**2 + cos(lat0) * cos(lat1) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    return R * c # return distance in km
```

```python
print(measure_distance(data_sampled['latitude'][19], data_sampled['longitude'][19], data_sampled['latitude'][372], data_sampled['longitude'][372]))
print(data.shape)
#print(data_sampled.shape)
```

```python
distance_matrix = np.zeros((data_sampled.shape[0], data_sampled.shape[0]))
for i in range(data_sampled.shape[0]):
    for j in range(data_sampled.shape[0]):
        distance_matrix[i, j] = measure_distance(data_sampled['latitude'][i],
                                                 data_sampled['longitude'][i],
                                                 data_sampled['latitude'][j],
                                                 data_sampled['longitude'][j])
        
```
