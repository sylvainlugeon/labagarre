```python
data = pd.read_csv('../data/AB_NYC_2019.csv')
```

```python
data.head()
```

```python
data_1000 = data.sample(n=1000)
```

```python
locations = data_1000.loc[:, ["neighbourhood_group",
                              "latitude",
                              "longitude",
                             "minimum_nights",
                             "price"]]
```

```python
for index, row in locations.iterrows():
    color="#E37222" # tangerine
    folium.CircleMarker(location=(row["latitude"],
                                  row["longitude"]),
                        radius=1,
                        color=color,
                        fill=True).add_to(folium_map)
```

```python
for index, row in locations.iterrows():
    if(row['neighbourhood_group'] == "Manhattan"):
        color="#E37222" # tangerine
    elif(row['neighbourhood_group'] == "Brooklyn"):
        color="#d022e3"
    elif(row['neighbourhood_group'] == "Queens"):
        color="#22e379"
    elif(row['neighbourhood_group'] == "Bronx"):
        color="#e32222"
    elif(row['neighbourhood_group'] == "Staten Island"):
        color="#5522e3"
    folium.CircleMarker(location=(row["latitude"],
                                  row["longitude"]),
                        radius=1,
                        color=color,
                        fill=True).add_to(folium_map)
```

```python
for index, row in locations.iterrows():
    if(row['minimum_nights'] == 1):
        color="#E37222" # tangerine
    elif(1< row['minimum_nights'] < 10):
        color="#d022e3"
    elif(10 < row['minimum_nights'] < 20):
        color="#22e379"
    elif(20< row['minimum_nights'] <100):
        color="#e32222"
    elif(100< row['minimum_nights'] < 366):
        color="#5522e3"
    folium.CircleMarker(location=(row["latitude"],
                                  row["longitude"]),
                        radius=1,
                        color=color,
                        fill=True).add_to(folium_map)
```

```python
folium_map
```

```python
def generateBaseMap(default_location=[40.693943, -73.985880], default_zoom_start=12):
    base_map = folium.Map(location=default_location, control_scale=True, zoom_start=default_zoom_start)
    return base_map
```

```python
base_map = generateBaseMap()
base_map
```

```python
from folium.plugins import HeatMap
data_1000__sup_nights = data_1000[data_1000.minimum_nights > data_1000.mean(0)["minimum_nights"]].copy()
base_map = generateBaseMap()
HeatMap(data=data_1000[['latitude', 'longitude', 'minimum_nights']].groupby(['latitude', 'longitude']).sum().reset_index().values.tolist(), radius=8, max_zoom=13).add_to(base_map)
base_map
```

```python
data_1000_sup_mean = data_1000[data_1000.price>data_1000.mean(0)["price"]].copy()
base_map = generateBaseMap()
HeatMap(data=data_1000[['latitude', 'longitude', 'price']].groupby(['latitude', 'longitude']).sum().reset_index().values.tolist(), radius=8, max_zoom=13).add_to(base_map)
#base_map.add_child(folium.ClickForMarker(popup='Potential Location'))
base_map
```
